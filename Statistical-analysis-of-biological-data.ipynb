<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

ومن الواضح جدا انه الارتباط بين الاعمدة ضعيف جدا يكاد لايذكر
</p>

<p style="font-family: Arial; font-size:1.75em;color:#09bef3; font-style:bold; text-align:right; direction: rtl; font-weight: bold" ><br>


اختبار ت لعينتين منفصلتين (Two-sample T-test):
</p>

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

من اجل القيام بأختبار ت لعينتين منفصلتين نحتاج الى استخدام الوظيفة ttest_ind من حزمة stats في مكتبة Scipy وعلى اساس قيمة p-value سنحدد ما اذا كان هناك فرق احصائي جوهري بين العينتين او لا <br> 
<br> 
وعلى سبيل المثال سوف نقوم بأنشاء العينتين التاليتين ثم سنقوم بأختبار ت عليها لمعرفة ما اذا كان هناك فرق احصائي جوهري بين هذين التجربتين او لا
</p>

```python
treatment1 = df[df["Virulence"] == 0.5]["ShannonDiversity"]
treatment2 = df[df["Virulence"] == 0.6]["ShannonDiversity"]

stats.ttest_ind(treatment1, treatment2)
```


<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

وبما ان p-value اكبر من 5% نستطيع ان نقول وبثقة انه ليس هناك فرق احصائي كبير بين التجربتين او العينتين
</p>

<p style="font-family: Arial; font-size:1.75em;color:#09bef3; font-style:bold; text-align:right; direction: rtl; font-weight: bold" ><br>


الاختبار (MWW RankSum test):
</p>

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

في الدورة تعلمنا عدة انواع للاختبارات مثل اختبار ت لعينة او عينتين منفصلتين او عينتين مرتبطتين ولكن جميع هذه الاختبارات تفترض ان بياناتنا موزعة توزيع طبيعي
<br> 
الاختبار Mann-Whitney-Wilcoxon (MWW) RankSum test لايفترض ان بيانتنا موزعة طبيعيا ولذلك هو مفيد جدا لتحديد اذا كان هناك اختلاف جوهري لتوزيعين مختلفين  
<br> 
ونستطيع القيام بهذا الاختبار من خلال استخدام الوظيفة ranksums في حزمة stats من مكتبة Scipy

<br>
ولفهم هذا الاختبار لنفترض اننا نريد معرفة اذا كان هناك فرق احصائي بين التجربتين التاليتين

</p>

```python
# select two treatment data sets from the parasite data
treatment1 = df[df["Virulence"] == 0.5]["ShannonDiversity"]
treatment2 = df[df["Virulence"] == 0.8]["ShannonDiversity"]
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

والان سنقوم بالاختبار Ranksum وعلى اساس p-value سنحدد اذا كان هناك فرق في توزيع التجربتين او التوزيعين متشابهين
</p>

```python
stats.ranksums(treatment1, treatment2)
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

وبما ان p-value اصغر بكثير من 5% نستطيع ان نقول وبثقة ان التوزيعين مختلفين جدا 
</p>

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

وربما في بيانات اخرى ستعمل عليها مستقبلا سيكون لديك توزيعين متشابهين ومثال على ذلك لنأخذ التجربتين التاليتين ونختبر هل هناك فرق بين توزيعهما
</p>

```python
treatment3 = df[df["Virulence"] == 0.8]["ShannonDiversity"]
treatment4 = df[df["Virulence"] == 0.9]["ShannonDiversity"]

stats.ranksums(treatment3, treatment4)
```

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

وكما هو واضح ان p-value اكبر من 5% هذا يعني ان توزيع التجربتين 3,4 لايختلف كثيرا 

<br>
للتوضيح: هذا يعني ان التغير بين القيمتين 0.8, 0.9 في العمود Virulence لم يؤثر كثيرا على تغير العمود ShannonDiversity

</p>


<p style="font-family: Arial; font-size:1.75em;color:#09bef3; font-style:bold; text-align:right; direction: rtl; font-weight: bold" ><br>


تحليل التباين (ANOVA):
</p>

<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

لقد تعلمنا سابقا ان تحليل التباين ANOVA يستخدم مع اكثر من عينتين وهو الحل الامثل لاختبار اذا كان هناك فرق احصائي بين عدد كبير من العينات

<br>

الان سوف نقوم بتجربة تحليل التباين على ثلاث تجارب من اجل اثبات انه ليس هناك فرق احصائي كبير بين التجارب الثلاث التالية

</p>

```python
treatment1 = df[df["Virulence"] == 0.7]["ShannonDiversity"]
treatment2 = df[df["Virulence"] == 0.8]["ShannonDiversity"]
treatment3 = df[df["Virulence"] == 0.9]["ShannonDiversity"]
```


<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

ومن اجل القيام بأختبار ANOVA نحتاج الى استخدام الوظيفة f_oneway من حزمة stats في مكتبة Scipy
</p>

```python
stats.f_oneway(treatment1, treatment2, treatment3)
```


<p style="font-family: Arial; font-size:1.2em;color:black; font-style:normal; text-align:right; direction: rtl"><br>

وبما ان p-value اكبر من 5% نستطيع ان نقول وبثقة انه ليس هناك فرق احصائي كبير بين التجارب الثلاث
</p>
